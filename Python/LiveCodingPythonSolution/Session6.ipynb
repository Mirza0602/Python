```python
import pandas as pd
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource,HoverTool, ColorBar
from bokeh.palettes import magma,viridis
from bokeh.models.mappers import LinearColorMapper
```

## Import Data
From the Data-folder read the file ``ExerciseCholesterin.csv`` into a pandas dataframe.

```python
choldata=pd.read_csv("ExerciseCholesterin1.csv",index_col=0)
choldata.head()
```

## Visual Correlation Analysis 

Visualize the correlations between cholesterin and exercise in a Bokeh plot. How do the two variables correlate?

```python
output_notebook()
```

```python
source = ColumnDataSource(choldata)
options = dict(plot_width=800, plot_height=500,
               tools="pan,wheel_zoom,box_zoom,box_select,lasso_select,reset")

p = figure(title="cholesterin vs. exercise", x_axis_label="exercise (h/week)", y_axis_label="cholesterin(mg/dl)", **options)
p.circle("exercise(h/week)","chol(mg/dl)", color="blue", size=8, alpha=0.6, source=source)
show(p)
```

Visualize the correlations between cholesterin and exercise in a Bokeh plot. Now, the agegroup shall be encoded in the color of the markers. How do the two variables correlate?

```python
mapper=LinearColorMapper(palette=viridis(20),low=choldata["agegroup"].min(),high=choldata["agegroup"].max())
```

```python
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
```

```python
source = ColumnDataSource(choldata)
options = dict(plot_width=800, plot_height=500,
               tools="pan,wheel_zoom,box_zoom,box_select,lasso_select,reset")

p = figure(title="cholesterin vs. exercise", x_axis_label="exercise (h/week)", y_axis_label="cholesterin(mg/dl)", **options)
p.circle("exercise(h/week)","chol(mg/dl)", color={'field': 'agegroup', 'transform': mapper}, size=8, alpha=0.9, source=source)
show(p)
```
