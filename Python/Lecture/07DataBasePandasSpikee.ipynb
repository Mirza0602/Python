```python
import psycopg2
import numpy as np
import json
import pandas as pd
pd.options.display.max_rows = 999
```

```python
with open('config.json') as f:
    conf = json.load(f)
```

```python
print conf['database']
```

```python
conn_str = "host={} dbname={} user={} password={}".format(conf["host"], conf["database"], conf["user"], conf["passw"])
conn = psycopg2.connect(conn_str)
#conn = psycopg2.connect(host=awsDB,dbname="hrv_web",user=awsUser,password=awsPw)
```

```python
df = pd.read_sql('select * from profiles_webuser', con=conn)
```
