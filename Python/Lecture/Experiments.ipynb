```python
from graphviz import Digraph
dot = Digraph()
top = Digraph()
down = Digraph()
top.graph_attr.update(rank="same")
down.graph_attr.update(rank="same")
top.node('A', 'If condition',shape="diamond")
top.node('B', 'Else',shape="diamond")
dot.node('C', 'Instructions',shape="box")
dot.node('D', 'Instructions',shape="box")
dot.node('In','',shape="point")
down.node('Out2','',shape="point")
down.node('Out','',shape="point")
#dot.node('A', '<f0> a | <f1> b | <f2> c | <f3> d| <f4> e ',shape="record",color="blue")
dot.edge('In','A')
top.edge('A', 'B')
dot.edge('A', 'C')
dot.edge('B', 'D')
dot.edge('C','Out')
dot.edge('D','Out2')
#dot.edge('Out2','Out')
dot.subgraph(top)
dot.subgraph(down)
dot
```
